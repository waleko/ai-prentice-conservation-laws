```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use("dark_background")

import ai_prentice_wasserstein

from time import time
```

```python
double_pendulum_data = np.load("trajectories/double_pendulum.npz")
data = double_pendulum_data["data"]
```

```python
ai_prentice = ai_prentice_wasserstein.DimensionalityPrentice("double pendulum")
```

```python
t_start = time()
ai_prentice.fit(data)
t_end = time()
print(f"time: {np.round(t_end - t_start, 3)}")
```

```python
print(f"Dimensionality of the double pendulum: {ai_prentice.dimensionality}")
```

```python
ai_prentice.show_results(plt, double_pendulum_data["params"][:, :1], "energy")
```

```python
np.savez("double_pendulum_with_embedding.npz", data=data, embedding=ai_prentice.embeddings[1])
```
