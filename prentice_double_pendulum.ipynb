```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use("dark_background")

import ai_prentice_wasserstein

from functools import partial
from time import time

import warnings
warnings.filterwarnings("ignore", message="using precomputed metric; inverse_transform will be unavailable")
```

```python
double_pendulum_data = np.load("trajectories/double_pendulum.npz")
data = double_pendulum_data["data"]
```

```python
Predictor = ai_prentice_wasserstein.DimensionalityPredictor("double pendulum")
```

```python
t_start = time()
Predictor.fit(data)
t_end = time()
print(f"time: {np.round(t_end - t_start, 3)}")
```

```python
print(f"Dimensionality of the double pendulum data: {Predictor.dimensionality}")
```

```python
Predictor.show_results(plt, double_pendulum_data["params"][:, :1], "energy")
```
