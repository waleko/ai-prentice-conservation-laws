```python
import numpy as np

import matplotlib.pyplot as plt
# plt.style.use("dark_background")

import ai_prentice_wasserstein

from time import time
```

```python
double_pendulum_le_data = np.load("trajectories/double_pendulum_low_energy.npz")
data = double_pendulum_le_data["data"]
```

```python
ai_prentice = ai_prentice_wasserstein.DimensionalityPrentice("double pendulum with low energies")
```

```python
t_start = time()
ai_prentice.fit(data)
t_end = time()
print(f"time: {np.round(t_end - t_start, 3)}")
```

```python
print(f"Dimensionality of the double pendulum with low energies: {ai_prentice.dimensionality}")
```

```python
ai_prentice.show_results(plt, double_pendulum_le_data["params"][:, 1:], ["E1", "E2"])
```

```python
np.savez("double_pendulum_low_energy_with_embedding.npz", data=data, embedding=ai_prentice.embeddings[2])
```

```python
fig, ax = plt.subplots()
ai_prentice._plot_scores(ax)
plt.savefig("dp_le_scores.pdf")
```

```python
fig, ax = plt.subplots()
ai_prentice._plot_score_diffs(ax)
plt.savefig("dp_le_diffs.pdf")
```

```python
fig, ax = plt.subplots()
ai_prentice._plot_2d(fig, ax, double_pendulum_le_data["params"].T[1], "E1")
plt.savefig("dp_le_E1.pdf")
```

```python
fig, ax = plt.subplots()
ai_prentice._plot_2d(fig, ax, double_pendulum_le_data["params"].T[2], "E2")
plt.savefig("dp_le_E2.pdf")
```
