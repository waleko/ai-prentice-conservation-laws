```python
import numpy as np

import matplotlib.pyplot as plt
# plt.style.use("dark_background")

import ai_prentice_wasserstein
import utils

from time import time
```

```python
harmonic_oscillator_data = np.load("trajectories/harmonic_oscillator.npz")
data = harmonic_oscillator_data["data"]
```

```python
ai_prentice = ai_prentice_wasserstein.DimensionalityPrentice("harmonic_oscillator")
```

```python
ai_prentice.fit(data)
```

```python
embedding = ai_prentice.embeddings[1]
energy = harmonic_oscillator_data["params"].T[0]
```

```python
utils.plot_scores(plt.gca(), ai_prentice.scores, ai_prentice.errors)
# plt.savefig("ho_scores.pdf")
plt.show()
```

```python
utils.plot_score_diffs(plt.gca(), ai_prentice.scores, ai_prentice.errors, ai_prentice.dimensionality, ai_prentice.threshold)
# plt.savefig("ho_diffs.pdf")
plt.show()
```

```python
utils.plot_embedding_vs_conserved_quantity(plt.gca(), embedding, energy, "energy")
# plt.savefig("ho_energy.pdf")
plt.show()
```
