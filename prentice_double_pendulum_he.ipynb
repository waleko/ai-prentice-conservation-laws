```python
import numpy as np

import matplotlib.pyplot as plt
# plt.style.use("dark_background")

import ai_prentice_wasserstein

from time import time
```

```python
double_pendulum_he_data = np.load("trajectories/double_pendulum_high_energy.npz")
data = double_pendulum_he_data["data"]
```

```python
ai_prentice = ai_prentice_wasserstein.DimensionalityPrentice("double pendulum with high energy")
```

```python
t_start = time()
ai_prentice.fit(data)
t_end = time()
print(f"time: {np.round(t_end - t_start, 3)}")
```

```python
print(f"Dimensionality of the double pendulum with high energy: {ai_prentice.dimensionality}")
```

```python
ai_prentice.show_results(plt, double_pendulum_he_data["params"][:, :1], ["energy"])
```

```python
np.savez("double_pendulum_high_energy_with_embedding.npz", data=data, embedding=ai_prentice.embeddings[1])
```

```python
fig, ax = plt.subplots()
ai_prentice._plot_scores(ax)
plt.savefig("dp_he_scores.pdf")
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ai_prentice._plot_score_diffs(ax)
plt.savefig("dp_he_diffs.pdf")
```

```python
fig, ax = plt.subplots()
ai_prentice._plot_embedding_vs_conserved_quantity(ax, double_pendulum_he_data["params"].T[0], "energy")
plt.savefig("dp_he_energy.pdf")
```
